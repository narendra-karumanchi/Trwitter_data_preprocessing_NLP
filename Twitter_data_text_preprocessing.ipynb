# HTML tags Removal

```python
# Below code will remove the HTML tags in the text if there are any
import re
def remove_html(text):
    pattern = re.compile('<.*?>')
    return pattern.sub("",text)
```

```python
column = column.apply(remove_html)
# to view the changes converting into list
review_data = column.tolist()
review_data
```

# URL removal

```python
def url_removal(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub("",text)
```

```python
import string
exclude = string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,"")
    return text
```

# Stop Word Removal

```python
pip install nltk
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word not in stopwords.words("english"):
            new_text.append(word)
    return "  ".join(new_text)
```

```python
column = column.apply(remove_stopwords)
# to view the changes converting into list
review_data = column.tolist()
review_data
```

```python
# removing emojis, here the context of emoji is removed
import re
def remove_emojis_manually(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251" 
                               "]+", flags=re.UNICODE)
    clean_text = emoji_pattern.sub(r'', text)
    return clean_text
```

```python
column2 = column.apply(remove_emojis_manually)
# to view the changes converting into list
review_data = column2.tolist()
review_data
```

# Stemming and Lemmatization

```python
# Stemming
from nltk.stem import PorterStemmer
def stemming(text):
    new_text=[]
    obj = PorterStemmer()
    for i in text:
        stem_word=[obj.stem(word) for word in i.split()]
        new_text.append(stem_word)
        
    return new_text
    
```

```python
column1 = column1.apply(stemming)

# to view the changes converting into list
review_data = column1.tolist()
review_data
```
